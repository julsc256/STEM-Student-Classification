### Convert letter grades to number grades and find the average grade for student in their middle school years:
- Define each letter grade to it's numerical value
- Total for each student += (current value + whatever you put on the right side of that)
- Count for each student
- Find the average grade (list) for each dividing total over count

#### Grade 7 - Average Grades

```python
list_of_7 = all_grades['Grade 7 Grades'].to_list()
```

```python
average_grades_7 = []
for student in list_of_7:
    grade_total = 0
    num_grades = 1
    for grade in student.split(","):
        if "A+" in grade:
            grade_total += 100
            num_grades += 1
        elif "A" in grade:
            grade_total += 96
            num_grades += 1
        elif "A-" in grade:
            grade_total += 92
            num_grades += 1
        elif "B+" in grade:
            grade_total += 89
            num_grades +=1
        elif "B" in grade:
            grade_total += 86
            num_grades +=1
        elif "B-" in grade:
            grade_total += 82
            num_grades +=1
        elif "C+" in grade:
            grade_total += 79
            num_grades +=1
        elif "C" in grade:
            grade_total += 76
            num_grades +=1
        elif "C-" in grade:
            grade_total += 72
            num_grades +=1
        elif "D+" in grade:
            grade_total += 69
            num_grades +=1
        elif "D" in grade:
            grade_total += 66
            num_grades +=1
        elif "D-" in grade:
            grade_total += 62
            num_grades +=1
        elif "F" in grade:
            grade_total += 51
            num_grades +=1
    average_grades_7.append(round(grade_total/num_grades))
```
